```python
# Import the OpenCV library
from google.colab.patches import cv2_imshow
import cv2
import numpy as np
```

```python
# Read the image file
image = cv2.imread('/content/test2.jpg')

# Check if the image is loaded properly
if image is None:
    print('Error: Image not found')
    exit()

# Display the image
cv2_imshow(image)

# Wait for a key press
cv2.waitKey(0)

# Close all windows
cv2.destroyAllWindows()
```

```python
resized_image = cv2.resize(image, (100, 90))
cv2_imshow(resized_image)
```

```python
 rotated_image = cv2.rotate(image, cv2.ROTATE_90_CLOCKWISE)
 cv2_imshow(rotated_image)
```

```python
 flipped_image = cv2.flip(image, 1)
 cv2_imshow(flipped_image)
```

```python
 kernel_size = 3
 blurred_image = cv2.GaussianBlur(image, (kernel_size, kernel_size), 0)
 cv2_imshow(blurred_image)
```

```python
 gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
 cv2_imshow(gray_image)
```

```python
 duplicated_image = np.copy(image)
 # Define coordinates for the rectangle on the duplicated image
 start_x, start_y = 50, 50  # Top-left corner
 end_x, end_y = 100, 100    # Bottom-right corner

 # Draw the rectangle on the duplicated image (green color with thickness 2)
 cv2.rectangle(duplicated_image, (start_x, start_y), (end_x, end_y), (0, 255, 0), 2)

 # Display the original and duplicated images with the rectangle
 cv2_imshow(image)
 cv2_imshow(duplicated_image)
```

```python
cv2.putText(duplicated_image, 'Hello, OpenCV!', (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 100, 0), 2)
```

```python

# Read the image
image = cv2.imread('/content/biili lab1.jpg')

# Check if the image is loaded properly
if image is None:
    print('Error: Image not found')
else:
    # Resize the image to half its original dimensions
    resized_image = cv2.resize(image, (0, 0), fx=0.5, fy=0.5)

    # Rotate the resized image 180 degrees clockwise
    rotated_image = cv2.rotate(resized_image, cv2.ROTATE_180)

    # Flip the rotated image vertically
    flipped_image = cv2.flip(rotated_image, 0)

    cv2_imshow(image)
    cv2_imshow(resized_image)
    cv2_imshow(rotated_image)
    cv2_imshow(flipped_image)

    # Wait for a key press and close all windows
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
image = cv2.imread('/content/biili lab1.jpg')

# Check if the image is loaded properly
if image is None:
    print('Error: Image not found')
else:
    # Blur the image using Gaussian blur with a kernel size of 5x5
    blurred_image = cv2.GaussianBlur(image, (5, 5), 0)

    # Convert the blurred image to grayscale
    grayscale_image = cv2.cvtColor(blurred_image, cv2.COLOR_BGR2GRAY)

    # Save the grayscale image
    cv2.imwrite('grayscale_image.jpg', grayscale_image)

    # Display the original, blurred, and grayscale images
    cv2_imshow(image)
    cv2_imshow(blurred_image)
    cv2_imshow(grayscale_image)

    # Wait for a key press and close all windows
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```
